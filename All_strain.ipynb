```python
import pandas as pd
sig_all_strain = pd.read_table('../MASTER_MQ_sig_all_strain.txt',index_col=0)
sig_all_strain
```

```python

org1 = 'Y_pestis_Yp4003'
org2 = 'Y_pestis_CO92'
bbh = pd.read_table('Y_pestis_Yp4003_and_Y_pestis_CO92_BBH.tab',index_col=org1)
sig_all_strain = sig_all_strain.join(bbh)
sig_all_strain['$Y_pestis_CO92_proteins'] = sig_all_strain['Y_pestis_CO92'].str.split('|').str.get(2)
sig_all_strain
sig_all_strain  = sig_all_strain.reset_index().set_index('$Y_pestis_CO92_proteins')
sig_all_strain.to_csv('significant_proteins_all_strain_stats.tab',sep='\t', index=True)
sig_all_strain
```

```python
import pandas as pd
sig_late = pd.read_excel('Sig_late_031318.xlsx',sheet_name=0,index_col=0)

org1 = 'Y_pestis_Yp4003'
org2 = 'Y_pestis_CO92'
bbh = pd.read_table('Y_pestis_Yp4003_and_Y_pestis_CO92_BBH.tab',index_col=org1)
sig_late = sig_late.join(bbh)
sig_late['$Y_pestis_CO92_proteins'] = sig_late['Y_pestis_CO92'].str.split('|').str.get(2)
sig_late  = sig_late.reset_index().set_index('$Y_pestis_CO92_proteins')
sig_late.to_csv('significant_proteins_late_stats.tab',sep='\t', index=True)
with open('significant_proteins_late_IDs.tab', 'w') as out:
    out.write('$Protein\n')
    out.write('\n'.join(sig_late.index))
```
